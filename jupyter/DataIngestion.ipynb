FAISS

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import OllamaEmbeddings
```

```python
df = TextLoader("/media/lang_chain/Storage/Documents/udemy/Datasets/donald.txt")
docs = df.load()
```

```python
text_splitter =CharacterTextSplitter(chunk_size=200,chunk_overlap=30)
data = text_splitter.split_documents(docs)
data
```

```python
#create embedding
embedding = OllamaEmbeddings(model="gemma:latest")
db =FAISS.from_documents(docs,embedding)
db
```

```python
##querying
query = "On what year jury in New York found Trump guilty  on 34 felony counts?"
docs = db.similarity_search(query)
docs
```

##using retriever

```python

retriever = db.as_retriever()
retriever.invoke(query)
```

Similarity search with score</br>
The FAISS method to get the similarity_search_with_score which allows you to get the document along with the distance score with respect to query. The retured distance is L2 distance. Therefore, a lower score is better and calculated using manhatten distance.

```python
docs_and_score = db.similarity_search_with_score(query)
docs_and_score
```

```python
#passing vector directly
embedding_vectors = embedding.embed_query(query)
```

```python
embedding_vectors
```

```python
docs_and_score = db.similarity_search_by_vector(embedding_vectors)
```

```python
docs_and_score
```

```python
#to save the vector db and loading it
db.save_local("faiss_indeex")
```

```python
new_db  =FAISS.load_local('faiss_indeex',embeddings,allow_dangerous_deserialization=True)
```

```python
docs = new_db.similarity_search(query)
```

```python
from langchain_core.prompts import ChatPromptTemplate
template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an expert AI Engineer. Provide me answer based on the question"),
        ("user", "{input}")
    ]
)

template
```
